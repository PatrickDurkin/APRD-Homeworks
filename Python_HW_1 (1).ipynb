```python
import pandas as pd
import arrow as arrow
import statsmodels.api as sm

DATA_URL = 'http://128.138.93.164/aprd6342/data/TravelPonyFacebook.csv'
df = pd.read_csv(DATA_URL)
```

```python
df.head()
```

```python
df['CPI'] = df['Amount Spent (USD)']/df['Impressions']
df.head()
```

```python
df['Start Date'] = pd.to_datetime(df['Start Date'])
df['Day of Week'] = df['Start Date'].dt.day_name()

df.groupby('Day of Week',as_index=False).agg({'CPI': 'sum'})
```

```python
print(df[['Amount Spent (USD)','Reach','Frequency','Unique Clicks','Page Likes']].corr())
```

```python
model1 = sm.OLS(df['Unique Clicks'], df['Reach']+df['Frequency']).fit()
model1.summary()

```

```python
model2 = sm.OLS(df['Unique Clicks'], df['Reach']).fit()
model2.summary()
```

```python
model3 = sm.OLS(df['Unique Clicks'], df['Frequency']).fit()
model3.summary()
```

```python
y = df['Unique Clicks']
x = df[['Reach','Frequency']]

model4 = sm.OLS(y, x).fit()
model4.summary()
```

1) The worst day of the week in terms of CPI is Friday at 2.056639. Saturday is the cheapest to generate impressions with 1.311720.


2) Unique Clicks has the strongest correlation with Amount Spent at .88. This means amount spent and unique clicks have a strong positive relationship.


3)The model indicates that reach more strongly predicts Unique Clicks it has a much larger T value and when run seperately the adjusted R squared is much higher. A Unique Click can only happen once so Frequency isnt as applicable.
